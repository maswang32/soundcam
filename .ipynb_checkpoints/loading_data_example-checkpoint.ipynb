```python
"""
This Jupyter notebook provides an example for loading data from the Treated Room.
"""

import datasets.dataset
import datasets.roomsetup
import os
import numpy as np
```

### Loading Dataset Attributes

```python
DATASET_PATH = "/viscam/projects/soundcam/datasets/human_rgbd/release/TreatedRoom_preprocessed/Human1" #<---Use the path in your directory
import datasets.treatedroom #


#Load dataset attributes - the treated room is called "darkroom"
dr = datasets.dataset.Dataset(datasets.roomsetup.RoomSetup(datasets.treatedroom.speaker_xyz,
                datasets.treatedroom.mic_xyzs_base,
                datasets.treatedroom.x_min,
                datasets.treatedroom.x_max,
                datasets.treatedroom.y_min,
                datasets.treatedroom.y_max,
                datasets.treatedroom.walls), DATASET_PATH)


print("Treated Room Dataset Loaded")
```

### Loading Dataset Files

```python
#Loading centroid (human xy positions)
centroid = np.load(os.path.join(DATASET_PATH, "centroid.npy"))
print("Shape of Centroid:")
print(centroid.shape)


#Loading Room Impulse Response (1000 human locations x 10 microphones x M time samples)
RIRs = np.load(os.path.join(DATASET_PATH, "deconvolved.npy"), mmap_mode='r')
print("Shape of RIRs:")
print(RIRs.shape)
```

### Plotting Dataset

```python
dr.room_setup.plot_room(centroid)
```
